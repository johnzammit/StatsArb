```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from binance.client import Client
client = Client(api_key, 
                secret_key,
                tld='us')
```

```python
all_tickers_info = client.get_all_tickers()
all_tickers_info
```

```python
len(all_tickers_info)
```

```python
tickers = [ticker["symbol"] for ticker in all_tickers_info]
tickers
```

```python
data = None
for i in range(1):
    symbol = tickers[i]
    print(symbol)
    klines = client.get_historical_klines(symbol, Client.KLINE_INTERVAL_1DAY, "1 Oct, 2018")
    data = pd.DataFrame(klines, columns = ['timestamp', 'open', 'high', 'low', 'close', 
                                           'volume', 'close_time', 'quote_av', 'trades', 
                                           'tb_base_av', 'tb_quote_av', 'ignore' ])
    print("------------------------------------------------------")
```

```python
data
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
# Plotting the time series
# plt.figure(figsize=(10, 6))
plt.plot(data['timestamp'], data['close'], label='Bitcoin Price')
plt.title('Bitcoin Price Time Series')
plt.xlabel('Timestamp')
plt.ylabel('Closing Price (USD)')
# plt.legend()
# plt.grid(True)
plt.show()
```
